```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# part-2 task-1
table = pd.read_csv('2017_jun_final.csv')

# part-2 task-2
table.head()
```

```python
# part-2 task-3

rows, columns = table.shape

print("Кількість рядків:", rows)
print("Кількість стовпців:", columns)
```

```python
# part-2 task-4

# Визначення типів стовпців за допомогою атрибуту dtypes
column_types = table.dtypes

# Виведення результатів
print("Типи стовпців:")
print(column_types)
```

```python
# part-1 task-5
missing_values = table.isnull().sum()

# # частку пропусків у кожній колонці
missing_percentage = missing_values/table.shape[0]

# результfn
result_df = pd.DataFrame({
    'Кількість пропусків': missing_values,
    'Частка пропусків (%)': missing_percentage
})
print(result_df)
```

```python
# part-1 task-6
table_v2 = table.drop(labels=["Специализация", "Университет", "Предметная.область", "Валюта", "cls"], axis=1)
```

```python
# part-1 task-7
missing_values = table_v2.isnull().sum()

# # частку пропусків у кожній колонці
missing_percentage = missing_values/table_v2.shape[0]

# результfn
result_df = pd.DataFrame({
    'Кількість пропусків': missing_values,
    'Частка пропусків (%)': missing_percentage
})
print(result_df)
```

```python
# part-1 task-8
table_v3 = table_v2.dropna()
table_v3
```

```python
# part-2 task-9

rows, columns = table_v3.shape

print("Кількість рядків:", rows)
print("Кількість стовпців:", columns)
```

```python
# part-2 task-10
python_data = table_v3[table_v3['Язык.программирования'] == 'Python']
python_data
```

```python
# part-2 task-11

rows, columns = python_data.shape

print("Кількість рядків:", rows)
print("Кількість стовпців:", columns)
```

```python
# part-2 task-12

grouped_table = python_data.groupby('Должность')
print(grouped_table.size())
print(grouped_table)
```

```python
# part-2 task-13

aggregated_data = grouped_table.agg({'Зарплата.в.месяц': ['min', 'max']})

# Перейменування стовпців для зручності
aggregated_data.columns = ['Мінімальна зарплата', 'Максимальна зарплата']

print(aggregated_data)
```

```python
# part-2 task-14

# Приклад функції fill_avg_salary
def fill_avg_salary(row):
    return row.mean()

update_table = aggregated_data.apply(fill_avg_salary, axis=1, result_type='expand')
update_table.name = "Середня зарплата"
final_update_table = pd.concat([aggregated_data, update_table], axis=1)
final_update_table
```
